```python
import pandas as pd
dataset = "/DataUkuranBaju.csv"

df = pd.read_csv(dataset)

df.head(20)
```

```python
print(df['Tinggi Badan (cm)'].isna().sum())

df['Tinggi Badan (cm)'].fillna('N', inplace=True)

print(df['Tinggi Badan (cm)'].isna().sum())
```

```python
print(df['Berat Badan (kg)'].isna().sum())

df['Berat Badan (kg)'].fillna('N', inplace=True)

print(df['Berat Badan (kg)'].isna().sum())
```

```python
df['Jenis Kelamin'].replace(['99'], ['Pria'], inplace=True)

print(df['Jenis Kelamin'].value_counts())
```
